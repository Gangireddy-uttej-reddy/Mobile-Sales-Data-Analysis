## Data Visualization

```python
brand_sales = df.groupby("mobile_brand")["units_sold"].sum()

brand_sales.plot(kind="bar")

plt.title("Total Sales by Mobile Brand")
plt.xlabel("Mobile Brand")
plt.ylabel("Units Sold")
plt.show()
```

### This barplot Shows the total sales  for each mobile brand

```python
problem_products = df.sort_values(["initial_stock","units_sold"], ascending=[False,True]).head(10)

problem_products.plot(x="mobile_model", y=["initial_stock","units_sold"], kind="bar")

plt.title("High Stock but Low Sales")
plt.ylabel("Units")
plt.show()
```

### This showes the highest stock but low sales 

```python
brand_sales = df.groupby("mobile_brand")["units_sold"].sum()

brand_sales.plot(kind="pie", autopct='%1.1f%%')

plt.title("Sales Share by Mobile Brand")
plt.ylabel("")
plt.show()
```

```python
year_sales = df.groupby("year")["units_sold"].sum()

year_sales.plot(kind="line", marker="o")

plt.title("Mobile Sales Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Units Sold")
plt.show()
```

### This lineplto show the mobile sales in year 

```python
ram_sales = df.groupby("ram_GB")["units_sold"].sum()

ram_sales.plot(kind="bar")

plt.title("RAM vs Sales")
plt.xlabel("RAM (GB)")
plt.ylabel("Units Sold")
plt.show()
```

```python
sns.histplot(data=df, x="battery_mAh", bins=5)

plt.title("Distribution of Battery Capacity")
plt.xlabel("Battery mAh")
plt.ylabel("Number of Phones")

plt.show()
```

```python
sns.kdeplot(data=df, x="price", fill=True)

plt.title("Price Distribution of Mobile Phones")
plt.xlabel("Price")
plt.ylabel("Density")

plt.show()
```

```python
store_revenue = df.groupby("store_city")["revenue"].sum()

store_revenue.plot(kind="barh")

plt.title("Revenue by Store Location")
plt.xlabel("Revenue")
plt.ylabel("Store City")
plt.show()
```

```python
df["price_range"] = pd.cut(df["price"],
                           bins=[0,20000,40000,60000,100000],
                           labels=["0-20k","20k-40k","40k-60k","60k+"])

price_sales = df.groupby("price_range")["units_sold"].sum().reset_index()

sns.lineplot(data=price_sales, x="price_range", y="units_sold", marker="o")

plt.title("Price Range vs Sales")
plt.xlabel("Price Range")
plt.ylabel("Units Sold")

plt.show()
```

```python
store_revenue = df.groupby("store_city")["revenue"].sum().reset_index()

sns.scatterplot(data=store_revenue, x="store_city", y="revenue", s=200)

plt.title("Revenue by Store Location")
plt.xlabel("Store City")
plt.ylabel("Revenue")

plt.show()
```

```python
processor_sales = df.groupby("processor")["units_sold"].sum().reset_index()

sns.scatterplot(data=processor_sales, x="processor", y="units_sold", s=200)

plt.title("Processor vs Sales")
plt.xlabel("Processor Type")
plt.ylabel("Units Sold")

plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

store_brand_sales = df.groupby(["store_city","mobile_brand"])["units_sold"].sum().reset_index()

sns.barplot(data=store_brand_sales, x="store_city", y="units_sold", hue="mobile_brand")

plt.title("Mobile Brand Sales in Each Store")
plt.xlabel("Store City")
plt.ylabel("Units Sold")

plt.legend(title="Mobile Brand")

plt.show()
```
